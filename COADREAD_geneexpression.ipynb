```python
gene_expression = pd.read_csv("C:/Users/juan_/OneDrive/Escritorio/Ironhack/Week9_final_project/coadread_tcga_pan_can_atlas_2018/coadread_tcga_pan_can_atlas_2018/data_mrna_seq_v2_rsem_zscores_ref_all_samples.txt", delimiter='\t')
gene_expression
```

Let's keep same 'submitter_id' format than in df_concat. So we remove '-01' in column names. Remove 'Entrez_Gene_Id' column

```python
column_names = [col[:-3] if col.endswith('-01') else col for col in gene_expression.columns]
gene_expression.columns = column_names
gene_expression = gene_expression.drop(columns='Entrez_Gene_Id')
gene_expression
```

Let's create a subset with top 50 mutated genes in CRC

```python
df_concat = pd.read_csv("df_concat_mut.csv")
top50_genes_list = [col.upper() for col in df_concat.columns[24:75]]
```

```python
df_concat
```

```python
top50_genes = gene_expression[gene_expression['Hugo_Symbol'].isin(top50_genes_list)]
top50_genes.head()
```

```python
melted_df = pd.melt(top50_genes, id_vars=['Hugo_Symbol'], var_name='submitter_id', value_name='expression')

pivoted_top50_expression = melted_df.pivot(index='submitter_id', columns='Hugo_Symbol', values='expression').reset_index()

# Remove the name of the index column
pivoted_top50_expression.columns.name = None

pivoted_top50_expression.columns = [col + '_exp' if col != 'submitter_id' else col for col in pivoted_top50_expression.columns]

pivoted_top50_expression
```

```python
melted_df
```

```python
melted_df.to_csv('melted_top50_expression.csv', index = False)
```

```python
pivoted_top50_expression.to_csv('pivoted_top50_expression.csv', index=False)
```

```python
df_final = pd.merge(df_concat, pivoted_top50_expression, on='submitter_id', how='left')
df_final.to_csv('df_final.csv', index=False)
```
