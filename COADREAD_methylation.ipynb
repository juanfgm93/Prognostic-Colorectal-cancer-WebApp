```python
import pandas as pd
```

```python
methylation_df = pd.read_csv("C:/Users/juan_/OneDrive/Escritorio/Ironhack/Week9_final_project/coadread_tcga_pan_can_atlas_2018/coadread_tcga_pan_can_atlas_2018/data_methylation_hm27_hm450_merged.txt", delimiter='\t')
methylation_df
```

Let's keep same 'submitter_id' format than in df_concat. So we remove '-01' in column names. Remove non-relevant columns

```python
column_names = [col[:-3] if col.endswith('-01') else col for col in methylation_df.columns]
methylation_df.columns = column_names
methylation_df = methylation_df.drop(columns=['ENTITY_STABLE_ID', 'TRANSCRIPT_ID','DESCRIPTION'])
methylation_df
```

Let's create a subset with top 50 mutated genes in CRC

```python
df_concat = pd.read_csv("df_concat_mut.csv")
top50_genes_list = [col.upper() for col in df_concat.columns[24:75]]
```

```python
top50_genes = methylation_df[methylation_df['NAME'].isin(top50_genes_list)]
top50_genes
```

Keep only one gene name

```python
top50_genes_mean = top50_genes.groupby('NAME').mean().reset_index()
top50_genes_mean
```

Which genes are lacking in top50_genes:

```python
genes_methilation = top50_genes['NAME'].tolist()

# Find genes in top50_genes dataframe but not in gene_expression dataframe
genes_not_in_expression = [gene for gene in top50_genes_list if gene not in genes_methilation]
genes_not_in_expression
```

```python
melted_df = pd.melt(top50_genes, id_vars=['NAME'], var_name='submitter_id', value_name='methilation')

# Using pivot_table instead of pivot to handle duplicate entries
pivoted_top50_methilation = melted_df.pivot_table(index='submitter_id', columns='NAME', values='methilation', aggfunc='mean').reset_index()

# Remove the name of the index column
pivoted_top50_methilation.columns.name = None

# Rename columns adding '_met' suffix
pivoted_top50_methilation.columns = [col + '_met' if col != 'submitter_id' else col for col in pivoted_top50_methilation.columns]

pivoted_top50_methilation
```

```python
pivoted_top50_methilation.to_csv('pivoted_top50_methilation.csv', index=False)
```

```python
melted_df.to_csv('melted_top50_methylation.csv', index = False)
```

```python
df_final = pd.read_csv("df_final.csv")
df_final2 = pd.merge(df_final, pivoted_top50_methilation, on='submitter_id', how='left')
df_final2.to_csv('df_final2.csv', index=False)
```
