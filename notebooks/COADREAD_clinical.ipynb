Let's check how many 'None' do we have in columns four, five and six

```python
def count_none_occurrences(column):
    none_count = 0
    for value in column:
        if value is None:
            none_count += 1
    return none_count

columns_to_count = ['samples_four', 'samples_five', 'samples_six']
none_counts = {}

# Iterate over the columns and count occurrences of None
for column_name in columns_to_count:
    none_counts[column_name] = count_none_occurrences(df_biospecimen_2[column_name])

# Display the counts of None for each column
for column_name, count in none_counts.items():
    print("Number of 'None' in '{}' column: {}".format(column_name, count))
```
